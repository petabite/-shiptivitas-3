```python
import time

before_release = []
after_release = []
with open('user-visits.txt', 'r') as f:
    timestamps = [int(i) for i in f.read().splitlines()] #read the lines from file created by SQL squery and convert to int
for timestamp in timestamps:
    if timestamp < 1527922800:
        # if epoch timestamp before release of app, append to before_release list
        before_release.append(timestamp)
    else:
        # else, append to after_release list
        after_release.append(timestamp)
        
print(len(before_release)) # print number of total users visits before release
print(len(set([time.strftime('%Y-%m-%d', time.localtime(timestamp)) for timestamp in before_release]))) #print total unique days that visitors visited the site on before release
print(len(after_release))  # print number of total users visits after release
print(len(set([time.strftime('%Y-%m-%d', time.localtime(timestamp)) for timestamp in after_release]))) #print total unique days that visitors visited the site on after release
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('Before feature change', 'After feature change')
y_pos = np.arange(len(objects))
user_visits = [493/112, 3567/245] # division to get average users per day
 
plt.bar(y_pos, user_visits, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Daily Average Users')
plt.title('Daily Average Users Before and After Feature Change')
 
plt.show() #plot graph below
```

```python
import collections

with open('status-changes.txt', 'r') as f:
    cards = [int(i) for i in f.read().splitlines()] #read the lines from file created by SQL squery and convert to int
print(cards)
counter = collections.Counter(cards) #init counter to count frequency of each cardID. Frequency represents the # of times the card was updated 
print(counter)
from pylab import *
t = counter.keys() # x-axis - each unique card id
s = counter.values() # y-axis - number of times card was updated
plot(t, s)
 
xlabel('Card ID')
ylabel('Number of status changes')
title('Number of status changes per unique Card ID')
grid(True)
show() # plot graph below
```
